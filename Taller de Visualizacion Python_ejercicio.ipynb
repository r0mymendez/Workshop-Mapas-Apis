1. Importar el paquete Folium

```
import folium
```

```
def mapa_markers(dataframe_,lat_,long_,popup_,lat_ini,long_ini,zoom_=11):
  map = folium.Map(location=[lat_ini, long_ini], zoom_start=11)
  for i in range(0, dataframe_.shape[0]):
    folium.Marker([dataframe_.loc[i,lat_], dataframe_.loc[i,long_]],popup=dataframe_.loc[0,'marca']).add_to(map)
  return map
```

```
def mapa_markers_colores_(dataframe_,lat_,long_,popup_,lat_ini,long_ini,zoom_,marca_1,color_1,marca_2,color_2):
  map = folium.Map(location=[lat_ini, long_ini], zoom_start=11)
  for i in range(0, dataframe_.shape[0]):
    if dataframe_.loc[i,'marca']==marca_1:
      folium.Marker([dataframe_.loc[i,lat_], dataframe_.loc[i,long_]],popup=dataframe_.loc[0,'marca'], icon=folium.Icon(color =color_1)).add_to(map)
    elif dataframe_.loc[i,'marca']==marca_2:
      folium.Marker([dataframe_.loc[i,lat_], dataframe_.loc[i,long_]],popup=dataframe_.loc[0,'marca'], icon=folium.Icon(color =color_2)).add_to(map)
    else:
      folium.Marker([dataframe_.loc[i,lat_], dataframe_.loc[i,long_]],popup=dataframe_.loc[0,'marca'], icon=folium.Icon(color =color_3)).add_to(map)
  return map
```
